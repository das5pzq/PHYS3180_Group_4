# Background Extraction
***
Extract the background from the spectral lines

```python
#Import Packages
##Import Packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy


from scipy.signal import find_peaks
```

```python
#Get Data
cols = ["wavelength", "intensity"]
raw_data = pd.read_csv("DataFiles/Hydrogen_Test_slit_bigger_300_900_nm_run_multi_1.txt", delimiter="\t", names=cols)
#raw_background1 = pd.read_csv("DataFiles/test_background.txt", delimiter="\t", names=cols)
#raw_background2 = pd.read_csv("DataFiles/test_background_0.3.txt", delimiter="\t", names=cols)
raw_background = pd.read_csv("DataFiles/Neon_Test_slit_2_300_900_nm_run_1.txt", delimiter="\t", names=cols)
```

```python
#Get Peaks & Plot
plt.figure(dpi=175)
plt.title("Data & Background: Intensity vs Wavelength")
plt.xlabel("Wavelength [nm]")
plt.ylabel("Intensity [V]")

peaks, properties = find_peaks(raw_data["intensity"], height=0, prominence=0.15, distance=30)#, threshold=12, distance=1)#, width=2)

plt.plot(raw_data["wavelength"], raw_data["intensity"], lw=.5, label="Hydrogen")
#plt.plot(raw_background1["wavelength"], raw_background1["intensity"]-0.5, lw=.5)
plt.plot(raw_background["wavelength"], raw_background["intensity"], lw=.5, label="Helium")
plt.plot(raw_data["wavelength"][peaks], raw_data["intensity"][peaks], "x")#, label=raw_data["wavelength"][peaks])

plt.ylim(0,.2)
plt.legend()
plt.tight_layout()

plt.savefig("Plots/Data_Background.pdf")
```
