```python
#Import Packages
##Import Packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy


from scipy.signal import find_peaks
```

```python
cols = ["wavelength", "intensity"]
raw_data = pd.read_csv("C:/PHYS 3180/PHYS3180_Group_4/Lab05/DataFiles/Air_Test_slit_bigger_300_900_nm_run_1.txt", delimiter="\t", names=cols)
raw_background = pd.read_csv("C:/PHYS 3180/PHYS3180_Group_4/Lab05/DataFiles/test_background_0.3.txt", delimiter="\t", names=cols)

# raw_data
```

```python
peaks, properties = find_peaks(raw_data["intensity"], height=.5, prominence=.1, distance=1)#, threshold=12, distance=1)#, width=2)

peaks

raw_data["wavelength"][peaks]
```

```python
plt.figure(dpi=175)
plt.title("Intensity vs Wavelength for Air, with bigger slit")
plt.xlabel("Wavelength [nm]")
plt.ylabel("Intensity [V]")
plt.plot(raw_data["wavelength"], raw_data["intensity"], lw=.5)
# plt.plot(raw_data["wavelength"][peaks], raw_data["intensity"][peaks], "x", label=raw_data["wavelength"][peaks])
plt.plot(raw_data["wavelength"][peaks], raw_data["intensity"][peaks], "x",label = 'Peaks')
plt.legend()
```
