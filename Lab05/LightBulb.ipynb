```python
#Import Packages
##Import Packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy

from scipy.signal import find_peaks
```

```python
#Data 
cols = ["wavelength", "intensity"]
raw_data1 = pd.read_csv("DataFiles/test_flourescent.txt", delimiter="\t", names=cols)
raw_data2 = pd.read_csv("DataFiles/test_flourescent_lamp.txt", delimiter="\t", names=cols)
```

```python
plt.figure(dpi=175)
plt.title("Intensity vs Wavelength for Phone Runs")
plt.xlabel("Wavelength [nm]")
plt.ylabel("Intensity [V]")

plt.plot(raw_data1["wavelength"], raw_data1["intensity"], lw=.5, label="Flourscent")
plt.plot(raw_data2["wavelength"], raw_data2["intensity"], lw=.5, label="Lamp")
plt.legend()

plt.show()
```
