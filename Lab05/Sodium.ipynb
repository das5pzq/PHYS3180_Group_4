```python
#Import Packages
##Import Packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy


from scipy.signal import find_peaks
```

```python
cols = ["wavelength", "intensity"]
raw_data = pd.read_csv("DataFiles/Sodium Lamp 1_300_900_nm.txt", delimiter="\t", names=cols)
raw_data2 = pd.read_csv("DataFiles/Refresh Bulb_Sodium_Chamber_1_300_900_nm run 3.txt", delimiter="\t", names=cols)
raw_data3 = pd.read_csv("DataFiles/Refresh Bulb_Sodium_Chamber_1_450_650_nm run 4.txt", delimiter="\t", names=cols)
raw_data4 = pd.read_csv("DataFiles/Refresh Bulb_Sodium_Chamber_1_450_650_nm run 5.txt", delimiter="\t", names=cols)
raw_data5 = pd.read_csv("DataFiles/Refresh Bulb_Sodium_Chamber_1_450_650_nm run 6.txt", delimiter="\t", names=cols)

raw_data4
```

```python
peaks, properties = find_peaks(raw_data["intensity"], height=0, prominence=.7, distance=30)#, threshold=12, distance=1)#, width=2)

peaks

raw_data["wavelength"][peaks]
```

```python
raw_data["wavelength"][peaks].values
```

```python
plt.figure(dpi=175)
plt.title("Intensity vs Wavelength for Hydrogen: Run 3")
plt.xlabel("Wavelength [nm]")
plt.ylabel("Intensity [V]")
plt.plot(raw_data["wavelength"], raw_data["intensity"], lw=.5)
plt.plot(raw_data["wavelength"][peaks], raw_data["intensity"][peaks], "x", label=raw_data["wavelength"][peaks])
plt.legend()
```

```python
plt.figure(dpi=175)
plt.title("Intensity vs Wavelength for Sodium: Run 3")
plt.xlabel("Wavelength [nm]")
plt.ylabel("Intensity [V]")
#plt.plot(raw_data2["wavelength"], raw_data2["intensity"]+.5, lw=.5)
plt.plot(raw_data3["wavelength"], raw_data3["intensity"]+.5, lw=.5)
plt.plot(raw_data4["wavelength"], raw_data4["intensity"], lw=.5)
#plt.plot(raw_data5["wavelength"], raw_data5["intensity"], lw=.5)
#plt.plot(raw_data6["wavelength"], raw_data6["intensity"], lw=.5)

plt.vlines(raw_data["wavelength"][peaks].values, 0.7, 1, color="red", lw=.5, alpha=0.7)

plt.xlim(500,630)
```
